```python
import pandas as pd
import json


data = pd.read_csv('./toy_dataset_label.csv',sep='\t', encoding='utf-8')
data = data.rename(columns={'FILE':'image','AUTHOR':'author', 'TECHNIQUE':'style'})
data = data[['author','style']]

data['author'] = [x.lower() for x in data['author']]
data['style'] = [x.lower().split(',')[0] for x in data['style']]
labels_auth = data['author'].unique()
labels_style = data['style'].unique()


label_json_auth = dict()
for i, label in enumerate(labels_auth):
    label_json_auth[i] = label

label_json_sty = dict()
for i, label in enumerate(labels_style):
    label_json_sty[i] = label

print(label_json_auth)
print(label_json_sty)
```

```python
with open('label_author.json', 'w') as f:
    json.dump(label_json_auth, f)
```

```python
with open('label_style.json', 'w') as f:
    json.dump(label_json_sty, f)
```
